```python
housing['strata'] = pd.cut(housing['median_income'],bins=[0,1.5,3.0,4.5,6.,np.inf],\
                           labels=['low','median-low','median','median-high','high'])
```

```python
housing['median_val'] = pd.cut(housing['median_house_value'],
                               bins=[0,100000,110000,200000,264725,np.inf],\
                               labels=['low','median-low','median','median-high','high'])
```

```python
housing.plot(kind='scatter',x='latitude',y='longitude',figsize=(25,15),
             grid=True,fontsize=15,alpha=0.2,c = 'median_house_value',
             colorbar=True,cmap='jet')

plt.title(label='Precio de casas en california',fontsize=30)
plt.xlabel('latitude', fontsize=18)
plt.ylabel('longitude', fontsize=18)


circ = plt.Circle((34,-118),0.8,color='black',fill=False)
circ2 = plt.Circle((37.5,-122),0.8,color='black',fill=False)

text1 = plt.text(32.3,-119.5,s='Areas con mayor\ndensidad y costo',fontsize=20)

ax = plt.gca()
ax.add_patch(circ)
ax.add_patch(circ2)

ax.plot()
plt.savefig('precio_casas_en_california.png')
plt.show()
```
